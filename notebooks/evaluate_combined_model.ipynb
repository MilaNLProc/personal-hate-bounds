# Implement the SepHeads + boundary combined model

__Objective:__ evaluate a model that takes the predictions from a SepHeads (fine-tuned) model and combines them with those of the boundary model, with the latter overriding the former based on a confidence score.

The above can be achieved just at the prediction level (without actually combining models), or at the full model level.

**Prediction level**

Inputs:
- Predictions from the SepHeads model.
- Predictions from the boundary model (for the same samples), with confidence scores.
- A confidence score threshold (hyperparameter - set by the user).

Output:
- The prediction from SepHeads, if the boundary model's confidence score for that sample is below the threshold, otherwise the prediction from the boundary model.

Implementation: a function.

**Model level**

Input:
- Input sample (text).
- Confidence threshold.

Output:
- Predictions, either from the SepHeads model or the boundary one, according to the same logic as above.

Implementation: new model, taking a fine-tuned SepHeads model and a confidence threshold as the input to the constructor, and performing a call to GPT internally.

```python
KUMAR_TRAINING_DATA_PATH = '/data1/moscato/personalised-hate-boundaries-data/models/sepheads_model_training_test_subsampling_2/training_data_subsampled.csv'
KUMAR_TEST_DATA_PATH = '/data1/moscato/personalised-hate-boundaries-data/models/sepheads_model_training_test_subsampling_2/test_data_subsampled.csv'

ANNOTATORS_DATA_PATH = '/data1/moscato/personalised-hate-boundaries-data/data/kumar_perspective_clean/annotators_data.csv'

SEPHEADS_PREDICTED_LOGITS_OUTPUT_PATH = '/data1/moscato/personalised-hate-boundaries-data/models/sepheads_model_training_test_subsampling_2/checkpoint-42903/predicted_logits_test.npy'
SEPHEADS_PREDICTED_TOXIC_SCORE_OUTPUT_PATH = '/data1/moscato/personalised-hate-boundaries-data/models/sepheads_model_training_test_subsampling_2/checkpoint-42903/predicted_toxic_score_test.npy'

BOUNDARY_MODEL_PREDS_PATH = '/data1/moscato/personalised-hate-boundaries-data/data/boundary_model_predictions/paul_kumar_fullpredictions.csv'
```

```python
# Load test data.
test_data = pd.read_csv(KUMAR_TEST_DATA_PATH)

# Load annotators data.
annotators_data = pd.read_csv(ANNOTATORS_DATA_PATH)

# Load SepHeads predictions.
predicted_logits = np.load(SEPHEADS_PREDICTED_LOGITS_OUTPUT_PATH)
sepheads_predicted_toxic_score = np.load(SEPHEADS_PREDICTED_TOXIC_SCORE_OUTPUT_PATH)

# Load boundary model predictions.
boundary_model_preds = pd.read_csv(BOUNDARY_MODEL_PREDS_PATH).rename(
    columns={'prediction': 'boundary_model_predicted_toxic_score'}
)

# Add SepHeads predictions to test data.
test_data['sepheads_predicted_toxic_score'] = sepheads_predicted_toxic_score
test_data['sepheads_predicted_logits_class_0'] = predicted_logits[:, 0]
test_data['sepheads_predicted_logits_class_1'] = predicted_logits[:, 1]

# Merge test data with boundary model predictions.
test_data = pd.merge(
    left=test_data,
    right=boundary_model_preds[['text_id', 'boundary_model_predicted_toxic_score', 'confidence_score']],
    on='text_id',
    how='left'
)

# Merge test data with annotators data (to get the extreme annotator label).
test_data = pd.merge(
    left=test_data,
    right=annotators_data[['annotator_id', 'extreme_annotator']],
    on='annotator_id',
    how='left'
)

# Compute majority-vote labels and add them to the data.
test_data = pd.merge(
    left=test_data,
    right=test_data.groupby('text_id')['toxic_score'].apply(
        lambda g: g.value_counts(ascending=False).index[0]
    ).rename('majority_vote_toxic_score').reset_index(),
    on='text_id',
    how='left'
)

test_data
```

### Evaluate the combined model

#### Metrics against annotator-level ground truth

```python
classification_metrics_kumar = []
classification_metrics_extreme_annotators_kumar = []
classification_metrics_nonextreme_annotators_kumar = []

for confidence_threshold in np.arange(70, 101, 5):
    # All annotators
    combined_preds = compute_combined_prediction(
        sepheads_preds=test_data['sepheads_predicted_toxic_score'],
        boundary_model_preds=test_data['boundary_model_predicted_toxic_score'],
        confidence_scores=test_data['confidence_score'],
        confidence_threshold=confidence_threshold
    )
    
    test_data[f'combined_predicted_toxic_score_confidence_{confidence_threshold}'] = combined_preds

    cr = classification_report(
        test_data['toxic_score'],
        combined_preds,
        digits=3,
        output_dict=True
    )

    overridden_preds_frac = (
        test_data[f'combined_predicted_toxic_score_confidence_{confidence_threshold}']
        != test_data['sepheads_predicted_toxic_score']
    ).mean()

    classification_metrics_kumar.append({
        'confidence_threshold': confidence_threshold,
        'overridden_preds_frac': overridden_preds_frac,
        'accuracy': cr['accuracy'],
        **cr['macro avg']
    })

    # Extreme annotators.
    cr = classification_report(
        test_data[test_data['extreme_annotator']]['toxic_score'],
        test_data[test_data['extreme_annotator']][f'combined_predicted_toxic_score_confidence_{confidence_threshold}'],
        digits=3,
        output_dict=True
    )

    overridden_preds_frac = (
        test_data[test_data['extreme_annotator']][f'combined_predicted_toxic_score_confidence_{confidence_threshold}']
        != test_data[test_data['extreme_annotator']]['sepheads_predicted_toxic_score']
    ).mean()

    classification_metrics_extreme_annotators_kumar.append({
        'confidence_threshold': confidence_threshold,
        'overridden_preds_frac': overridden_preds_frac,
        'accuracy': cr['accuracy'],
        **cr['macro avg']
    })

    # Non-extreme annotators.
    cr = classification_report(
        test_data[~test_data['extreme_annotator']]['toxic_score'],
        test_data[~test_data['extreme_annotator']][f'combined_predicted_toxic_score_confidence_{confidence_threshold}'],
        digits=3,
        output_dict=True
    )

    overridden_preds_frac = (
        test_data[~test_data['extreme_annotator']][f'combined_predicted_toxic_score_confidence_{confidence_threshold}']
        != test_data[~test_data['extreme_annotator']]['sepheads_predicted_toxic_score']
    ).mean()

    classification_metrics_nonextreme_annotators_kumar.append({
        'confidence_threshold': confidence_threshold,
        'overridden_preds_frac': overridden_preds_frac,
        'accuracy': cr['accuracy'],
        **cr['macro avg']
    })

classification_metrics_kumar = pd.DataFrame(classification_metrics_kumar)
classification_metrics_extreme_annotators_kumar = pd.DataFrame(classification_metrics_extreme_annotators_kumar)
classification_metrics_nonextreme_annotators_kumar = pd.DataFrame(classification_metrics_nonextreme_annotators_kumar)


# Baseline.
classification_metrics_kumar_baselines = []

cr = classification_report(
    test_data['toxic_score'],
    test_data['sepheads_predicted_toxic_score'],
    digits=3,
    output_dict=True
)

classification_metrics_kumar_baselines.append({
    'group': 'all',
    'accuracy': cr['accuracy'],
    **cr['macro avg']
})

# Extreme annotators.
cr = classification_report(
    test_data[test_data['extreme_annotator']]['toxic_score'],
    test_data[test_data['extreme_annotator']]['sepheads_predicted_toxic_score'],
    digits=3,
    output_dict=True
)

classification_metrics_kumar_baselines.append({
    'group': 'extreme_annotators',
    'accuracy': cr['accuracy'],
    **cr['macro avg']
})

# Non-extreme annotators.
cr = classification_report(
    test_data[~test_data['extreme_annotator']]['toxic_score'],
    test_data[~test_data['extreme_annotator']]['sepheads_predicted_toxic_score'],
    digits=3,
    output_dict=True
)

classification_metrics_kumar_baselines.append({
    'group': 'nonextreme_annotators',
    'accuracy': cr['accuracy'],
    **cr['macro avg']
})

classification_metrics_kumar_baselines = pd.DataFrame(classification_metrics_kumar_baselines)
```

```python
kumar_annotator_welfare = test_data.groupby('annotator_id').agg(
    n_preds_total=pd.NamedAgg('text_id', 'count'),
    extreme_annotator=pd.NamedAgg('extreme_annotator', 'first')
)

confidence_thresholds = sorted([int(c.split('_')[-1]) for c in test_data.columns if 'combined' in c])

for confidence_threshold in confidence_thresholds[::-1]:
    print(f'Processing confidence threshold: {confidence_threshold}')

    annotator_welfare_confidence_threshold = pd.merge(
        left=test_data.groupby('annotator_id')[['sepheads_predicted_toxic_score', f'combined_predicted_toxic_score_confidence_{confidence_threshold}']].apply(
            lambda g: (g['sepheads_predicted_toxic_score'] != g[f'combined_predicted_toxic_score_confidence_{confidence_threshold}']).sum()
        ).reset_index().rename(columns={0: f'n_overridden_preds_confidence_{confidence_threshold}'}),
        right=test_data.groupby('annotator_id')[['toxic_score', f'combined_predicted_toxic_score_confidence_{confidence_threshold}']].apply(
            lambda g: classification_report(y_true=g['toxic_score'], y_pred=g[f'combined_predicted_toxic_score_confidence_{confidence_threshold}'], output_dict=True, zero_division=np.nan)
        ).rename(f'annotator_cr_confidence_{confidence_threshold}').reset_index(),
        # right=test_data.groupby('annotator_id')[['toxic_score', f'combined_predicted_toxic_score_confidence_{confidence_threshold}']].apply(
        #     lambda g: accuracy_score(
        #         y_true=g['toxic_score'],
        #         y_pred=g[f'combined_predicted_toxic_score_confidence_{confidence_threshold}'],
        #     )
        # ).rename(f'accuracy_confidence_{confidence_threshold}').reset_index(),
        how='left',
        on='annotator_id'
    )

    kumar_annotator_welfare = pd.merge(
        left=kumar_annotator_welfare,
        right=annotator_welfare_confidence_threshold,
        how='left',
        on='annotator_id'
    )

kumar_annotator_welfare = pd.merge(
    left=kumar_annotator_welfare,
    right=test_data.groupby('annotator_id')[['toxic_score', 'sepheads_predicted_toxic_score']].apply(
        lambda g: classification_report(y_true=g['toxic_score'], y_pred=g['sepheads_predicted_toxic_score'], output_dict=True, zero_division=np.nan)
    ).rename('annotator_cr_sepheads').reset_index(),
    # right=test_data.groupby('annotator_id')[['toxic_score', 'sepheads_predicted_toxic_score']].apply(
    #     lambda g: accuracy_score(y_true=g['toxic_score'], y_pred=g['sepheads_predicted_toxic_score'])
    # ).rename('accuracy_sepheads').reset_index(),
    how='left',
    on='annotator_id'
)

# Extract F1 scores in separate columns.
for ct in confidence_thresholds:
    kumar_annotator_welfare[f'macro_f1_confidence_{ct}'] = kumar_annotator_welfare[f'annotator_cr_confidence_{ct}'].apply(lambda cr: cr['macro avg']['f1-score'])

kumar_annotator_welfare['macro_f1_sepheads'] = kumar_annotator_welfare[f'annotator_cr_sepheads'].apply(lambda cr: cr['macro avg']['f1-score'])

kumar_annotator_welfare
```

```python
# Load dataset.
hatecheck_data = datasets.load_dataset(
    path="Paul/hatecheck",
    cache_dir='/data1/shared_datasets/'
)['test'].to_pandas()

hatecheck_data['label_gold'] = hatecheck_data['label_gold'].map({
    'hateful': 1,
    'non-hateful': 0
})

# Attach predictions from the boundary model.
hatecheck_data = pd.concat(
    [
        hatecheck_data,
        pd.read_csv(BOUNDARY_MODEL_PREDS_HATECHECK_PATH)
    ],
    axis=1
).rename(columns={'class_label': 'boundary_model_predicted_toxic_score'})

# Correctly map the predictions from the boundary model.
hatecheck_data['boundary_model_predicted_toxic_score'] = hatecheck_data['boundary_model_predicted_toxic_score'].map({
    '1': 1,
    'yes': 1,
    'no': 0
})

# Merge data and predictions from both models.
hatecheck_data = pd.merge(
    left=sepheads_hatecheck_preds_df,
    right=hatecheck_data,
    on='case_id',
    how='left'
)

# Join with annotators data.
hatecheck_data = pd.merge(
    left=hatecheck_data,
    right=annotators_data[['annotator_id', 'extreme_annotator']],
    how='left',
    on='annotator_id'
)
```

### Evaluate combined model on HateCheck

#### Global metrics

```python
classification_metrics_hatecheck = []
classification_metrics_extreme_annotators_hatecheck = []
classification_metrics_nonextreme_annotators_hatecheck = []

for confidence_threshold in np.arange(70, 101, 5):
    # All annotators.
    combined_preds = compute_combined_prediction(
        sepheads_preds=hatecheck_data['sepheads_predicted_toxic_score'],
        boundary_model_preds=hatecheck_data['boundary_model_predicted_toxic_score'],
        confidence_scores=hatecheck_data['confidence_score'],
        confidence_threshold=confidence_threshold
    )
    
    hatecheck_data[f'combined_predicted_toxic_score_confidence_{confidence_threshold}'] = combined_preds

    cr = classification_report(
        hatecheck_data['label_gold'],
        combined_preds,
        digits=3,
        output_dict=True
    )

    overridden_preds_frac = (
        hatecheck_data[f'combined_predicted_toxic_score_confidence_{confidence_threshold}']
        != hatecheck_data['sepheads_predicted_toxic_score']
    ).mean()

    classification_metrics_hatecheck.append({
        'confidence_threshold': confidence_threshold,
        'overridden_preds_frac': overridden_preds_frac,
        'accuracy': cr['accuracy'],
        **cr['macro avg']
    })

    # Extreme annotators.
    cr = classification_report(
        hatecheck_data[hatecheck_data['extreme_annotator']]['label_gold'],
        hatecheck_data[hatecheck_data['extreme_annotator']][f'combined_predicted_toxic_score_confidence_{confidence_threshold}'],
        digits=3,
        output_dict=True
    )

    overridden_preds_frac = (
        hatecheck_data[hatecheck_data['extreme_annotator']][f'combined_predicted_toxic_score_confidence_{confidence_threshold}']
        != hatecheck_data[hatecheck_data['extreme_annotator']]['sepheads_predicted_toxic_score']
    ).mean()

    classification_metrics_extreme_annotators_hatecheck.append({
        'confidence_threshold': confidence_threshold,
        'overridden_preds_frac': overridden_preds_frac,
        'accuracy': cr['accuracy'],
        **cr['macro avg']
    })

    # Non-extreme annotators.
    cr = classification_report(
        hatecheck_data[~hatecheck_data['extreme_annotator']]['label_gold'],
        hatecheck_data[~hatecheck_data['extreme_annotator']][f'combined_predicted_toxic_score_confidence_{confidence_threshold}'],
        digits=3,
        output_dict=True
    )

    overridden_preds_frac = (
        hatecheck_data[~hatecheck_data['extreme_annotator']][f'combined_predicted_toxic_score_confidence_{confidence_threshold}']
        != hatecheck_data[~hatecheck_data['extreme_annotator']]['sepheads_predicted_toxic_score']
    ).mean()

    classification_metrics_nonextreme_annotators_hatecheck.append({
        'confidence_threshold': confidence_threshold,
        'overridden_preds_frac': overridden_preds_frac,
        'accuracy': cr['accuracy'],
        **cr['macro avg']
    })
    
classification_metrics_hatecheck = pd.DataFrame(classification_metrics_hatecheck)
classification_metrics_extreme_annotators_hatecheck = pd.DataFrame(classification_metrics_extreme_annotators_hatecheck)
classification_metrics_nonextreme_annotators_hatecheck = pd.DataFrame(classification_metrics_nonextreme_annotators_hatecheck)
```

## Annotator welfare analysis

Analyses (con the predictions from the combined models):
- Number of annotators for which at least one prediction is overridden by the boundary model, as a function of the confidence threshold.
- Evaluation of annotator welfare at the annotator level (grouping by annotator).

```python
hatecheck_annotator_welfare = hatecheck_data.groupby('annotator_id').agg(
    n_preds_total=pd.NamedAgg('case_id', 'count'),
    extreme_annotator=pd.NamedAgg('extreme_annotator', 'first')
)

confidence_thresholds = sorted([int(c.split('_')[-1]) for c in hatecheck_data.columns if 'combined' in c])

for confidence_threshold in confidence_thresholds[::-1]:
    print(f'Processing confidence threshold: {confidence_threshold}')

    annotator_welfare_confidence_threshold = pd.merge(
        left=hatecheck_data.groupby('annotator_id')[['sepheads_predicted_toxic_score', f'combined_predicted_toxic_score_confidence_{confidence_threshold}']].apply(
            lambda g: (g['sepheads_predicted_toxic_score'] != g[f'combined_predicted_toxic_score_confidence_{confidence_threshold}']).sum()
        ).reset_index().rename(columns={0: f'n_overridden_preds_confidence_{confidence_threshold}'}),
        right=hatecheck_data.groupby('annotator_id')[['label_gold', f'combined_predicted_toxic_score_confidence_{confidence_threshold}']].apply(
            lambda g: classification_report(y_true=g['label_gold'], y_pred=g[f'combined_predicted_toxic_score_confidence_{confidence_threshold}'], output_dict=True, zero_division=0.0)
        ).rename(f'annotator_cr_confidence_{confidence_threshold}').reset_index(),
        how='left',
        on='annotator_id'
    )

    hatecheck_annotator_welfare = pd.merge(
        left=hatecheck_annotator_welfare,
        right=annotator_welfare_confidence_threshold,
        how='left',
        on='annotator_id'
    )

hatecheck_annotator_welfare = pd.merge(
    left=hatecheck_annotator_welfare,
    right=hatecheck_data.groupby('annotator_id')[['label_gold', 'sepheads_predicted_toxic_score']].apply(
        lambda g: classification_report(y_true=g['label_gold'], y_pred=g['sepheads_predicted_toxic_score'], output_dict=True, zero_division=0.0)
    ).rename('annotator_cr_sepheads').reset_index(),
    how='left',
    on='annotator_id'
)

for ct in confidence_thresholds:
    hatecheck_annotator_welfare[f'macro_f1_confidence_{ct}'] = hatecheck_annotator_welfare[f'annotator_cr_confidence_{ct}'].apply(lambda cr: cr['macro avg']['f1-score'])

hatecheck_annotator_welfare['macro_f1_sepheads'] = hatecheck_annotator_welfare[f'annotator_cr_sepheads'].apply(lambda cr: cr['macro avg']['f1-score'])
```

```python
# Table: confidence threshold, number of overridden annotators, number of annotators improving, number of annotators getting worse.
hatecheck_annotator_welfare_whole_population = []

for ct in confidence_thresholds[::-1]:
    increasing_f1s = hatecheck_annotator_welfare[
        hatecheck_annotator_welfare[f'macro_f1_confidence_{ct}'] > hatecheck_annotator_welfare['macro_f1_sepheads']
    ].reset_index(drop=True)
    decreasing_f1s = hatecheck_annotator_welfare[
        hatecheck_annotator_welfare[f'macro_f1_confidence_{ct}'] < hatecheck_annotator_welfare['macro_f1_sepheads']
    ].reset_index(drop=True)
    
    hatecheck_annotator_welfare_whole_population.append({
        'confidence_threshold': ct,
        'perc_annotators_with_overrides': round(
            100. * (hatecheck_annotator_welfare[f'n_overridden_preds_confidence_{ct}'] >= 1).sum() / len(hatecheck_annotator_welfare),
            2
        ),
        'perc_annotators_improving': (
            round(
                100. * (hatecheck_annotator_welfare[f'macro_f1_confidence_{ct}'] > hatecheck_annotator_welfare['macro_f1_sepheads']).sum() / len(hatecheck_annotator_welfare),
                2
            )
        ),
        'perc_annotators_getting_worse': (
            round(
                100. * (hatecheck_annotator_welfare[f'macro_f1_confidence_{ct}'] < hatecheck_annotator_welfare['macro_f1_sepheads']).sum() / len(hatecheck_annotator_welfare),
                2
            )
        ),
        'avg_macro_f1_delta_increase_median': round((increasing_f1s[f'macro_f1_confidence_{ct}'] - increasing_f1s[f'macro_f1_sepheads']).median(), 2),
        'avg_macro_f1_delta_decrease_median': round((decreasing_f1s[f'macro_f1_confidence_{ct}'] - decreasing_f1s[f'macro_f1_sepheads']).median(), 2)
    })

hatecheck_annotator_welfare_whole_population = pd.DataFrame(hatecheck_annotator_welfare_whole_population)

hatecheck_annotator_welfare_whole_population['n_annotators_total'] = len(hatecheck_annotator_welfare)

selected_confidence_thresholds = [100, 95, 90, 85, 80, 75, 70]

# To do: put in the paper.
hatecheck_annotator_welfare_whole_population_selected = hatecheck_annotator_welfare_whole_population[
    hatecheck_annotator_welfare_whole_population['confidence_threshold'].isin(selected_confidence_thresholds)
].reset_index(drop=True).drop(columns=['n_annotators_total'])

hatecheck_annotator_welfare_whole_population_selected
```
